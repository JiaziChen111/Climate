### Preliminary work: Importing required packages, importing parameters setting, defining functions

Recall that in order to solve the prefernce HJB model, we transform the question in solving a linear system of equations.
\begin{eqnarray}
{\frac {V(x) - {\widetilde V}(x)} \epsilon}  & =  &   V(x) \cdot A(x)+{\frac {\partial V}{\partial x}} (x) \cdot  B(x)   + {\frac {\partial^2 V}{\partial x \partial x'}}(x) \cdot C(x)+ D(x) 
\end{eqnarray}
Where
\begin{eqnarray}
A(x) &=& - \delta \\
B(x) &=& {\widehat \mu}(x)\\
C(x) &=& {\frac 1 2} \textrm{trace} \left[\sigma_X(x)' I \sigma_X(x) \right] \\
D(x) &=& \delta (1 - \kappa) \left( \log \left[ {\alpha}  - {\widehat i}(x)  - {\widehat j} (x)  \right] + k -  d  \right) + \delta \kappa \left[ \log {\widehat e}(x)   +  r \right] + {\frac {\xi_m} 2} {\widehat g}(x) \cdot {\widehat g}(x)   + \xi_p {\widehat {\mathbb I}}(x) \\
\end{eqnarray}

Precisely speaking our value function is a three-dimension tensor $ V_{initial}(R,T,K) $ with default grid size of (30,40,25) in "F" order. In order to solve the linear system, we reshape $ V(x) $ into a one dimension array with size of (30000,1) and conduct similar adjustments to $ A(x) $, $ B(x) $, $ C(x) $ and $ D(x) $ respectively.

In "F" order, when we call our initial value function at grid point (i,j,k) $ V_{initial}(i,j,k) $, we instead call equivalent reshaped value fnction $V(i + j * N_i + k * N_i * N_j)$ where $N_i$, $N_j$ and $N_k$ stands for number of grid poitns in that dimension.

Hence we apply forward difference for the first order derivative as follow, at upper boundary we use backward difference:
\begin{align*}
\frac{\partial V}{\partial R} (i,j,k) & =  \frac{V_{initial}(i,j,k)-V_{initial}(i-1,j,k)}{\Delta_{R}} = \frac{V(i + j * N_i + k * N_i * N_j)-V(i - 1 + j * N_i + k * N_i * N_j)}{\Delta_{R}}\\
\frac{\partial V}{\partial T} (i,j,k) & = \frac{V_{initial}(i,j,k)-V_{initial}(i,j-1,k)}{\Delta_{T}} = \frac{V(i + j * N_i + k * N_i * N_j)-V(i + (j-1) * N_i + k * N_i * N_j)}{\Delta_{T}}\\
\frac{\partial V}{\partial K} (i,j,k) & =  \frac{V_{initial}(i,j,k)-V_{initial}(i,j,k-1)}{\Delta_{K}} = \frac{V(i + j * N_i + k * N_i * N_j)-V(i + j * N_i + (k-1) * N_i * N_j)}{\Delta_{K}}\\
\end{align*}

For second order derivative, we use second order central difference scheme. While in the upper and lower bound, we apply second order backward and forward difference.
\begin{eqnarray*}
{\frac {\partial^2 V}{\partial R \partial R'}}(i,j,k) & = & \frac{V(i+1 + j * N_i + k * N_i * N_j)+V(i-1 + j * N_i + k * N_i * N_j)-2V(i + j * N_i + k * N_i * N_j)}{\Delta_{R}^{2}}\\
{\frac {\partial^2 V}{\partial T \partial T'}}(i,j,k) & = & \frac{V(i + (j+1) * N_i + k * N_i * N_j)+V(i + (j-1) * N_i + k * N_i * N_j)-2V(i + j * N_i + k * N_i * N_j)}{\Delta_{T}^{2}}\\
{\frac {\partial^2 V}{\partial K \partial K'}}(i,j,k) & = & \frac{V(i + j * N_i + (k+1) * N_i * N_j)+V(i-1 + j * N_i + (k-1) * N_i * N_j)-2V(i+1 + j * N_i + k * N_i * N_j)}{\Delta_{K}^{2}}\\
\end{eqnarray*}

We then will put all these derivatives in orginal HJB equation to form the linear system of equations.

So in that case, taking default grid size of (30,40,25) for $V_{initial}(5,10,20)$ i.e. $V(24305)$, its value function was used for computing for first and second derivatives of $V(24304)$, $V(24305)$ and $V(24306)$ with repect to the first dimension $R$, then $V(24335)$ and $V(24275)$ for first derivative in second dimension $T$, $V(23105)$, $V(25505)$ for first and second derivative in third dimension $K$. In our setting, the parameters for $V(24275)$ is 0 as we always set the second derivative with regard to T to 0. So we in total have 6 entries for each line in the 30,000 * 30,000 grid matrix
